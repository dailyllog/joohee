[문제2]
a = '14 push 1 push 2 top size empty'

a 문자열을 띄어쓰기로 나눈 후 리스트로 만들어서
push가 몇 개 있는지 찾으세용

```python
a =  '14 push 1 push 2 top size empty'

str = a.split(" ")
#print(type(str))
print(str.count('push'))
```

```python
a =  '14 push 1 push 2 top size empty'
lst_s = a.split(" ")
print(lst_s, type(lst_s))

sum = 0 
for i in lst_s:
    if 'push' in i:
        sum += 1

print(sum)
```

[문제3]
push가 2개인 경우가 있는지 없는지, 있으면 몇 개 있는지

```python
a = '14 pushpush 1 pushTopush 2 top push size empty'

a_list = a.split(' ')
#print(a_list)

sum = 0 
for i in a_list:
    if 'push' in i:
        print('push 있다!:')
        c = i.count('push')
        print('push 개수 :', c)
        sum += c
print(sum)
```
